```python
# Function to convert pdf to text file
from PyPDF2 import PdfReader

def pdf_to_text_pypdf2(pdf_path, output_txt_path):
    """
    Converts a PDF file to a plain text file using PyPDF2.

    Args:
        pdf_path (str): The path to the input PDF file.
        output_txt_path (str): The path to the output text file.
    """
    reader = PdfReader(pdf_path)
    with open(output_txt_path, "w", encoding="utf-8") as f:
        for page in reader.pages:
            text = page.extract_text()
            if text:
                f.write(text + '\n')
    print(f"Text extracted from '{pdf_path}' and saved to '{output_txt_path}' using PyPDF2.")

# Example usage:
# pdf_to_text_pypdf2("input.pdf", "output_pypdf2.txt")
```

```python
# Function to find the word frequencies from the text file
from collections import Counter
import re

def count_word_frequency(filepath):
    """Counts the frequency of words in a given text file."""
    try:
        with open(filepath, 'r', encoding='utf-8') as file:
            text = file.read()
            # Convert to lowercase and remove non-alphanumeric characters (keeping spaces)
            cleaned_text = re.sub(r'[^a-zA-Z\s]', '', text).lower()
            words = cleaned_text.split()
            word_counts = Counter(words)
            return word_counts
    except FileNotFoundError:
        print(f"Error: File not found at {filepath}")
        return None

# Example usage:
# file_path = "your_text_file.txt"  # Replace with the actual path to your text file
# word_frequencies = count_word_frequency(file_path)

# if word_frequencies:
#     print("Word frequencies:")
#     for word, count in word_frequencies.most_common(10):  # Print top 10 most common words
#         print(f"'{word}': {count}")
```

```python
# Give the file path where it can be accessed again in this ipynb
file_path = "/content/ArtOfWar.txt"
word_frequencies = count_word_frequency(file_path)
```
